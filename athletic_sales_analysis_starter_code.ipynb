### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
Total_products_sold_2020 = athletic_df_2020.groupby(['region', 'state', 'city']).sum()
Total_products_sold_2020.head(5)


# Show the top 5 results.
Total_products_sold_2020.head(5)                 
```

```python
Total_products_sold_2021 = athletic_df_2021.groupby(['region', 'state', 'city']).sum()
Total_products_sold_2021.head(5)
#i know i dont have to put 5 in the parenthesis at head, but it helps my ape brain. 
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
Total_sales_pivot_2020 = pd.pivot_table(Total_products_sold_2020, 
                                            index=['region','state','city'],
                                            values='total_sales', sort=True)
                                        
Total_sales_pivot_2020
# Rename the "units_sold" column to "Total_Products_Sold"
Total_sales_pivot_2020 = Total_sales_pivot_2020.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
Total_sales_pivot_2020.head(5)
```
